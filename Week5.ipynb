# Getting the co-ordinates of Karnataka using geopy

```python
address = 'Karnataka'

geoloc = Nominatim(user_agent="india")
try:
    location = geoloc.geocode(address, timeout=10)
    lat = location.latitude
    lon = location.longitude
    print('The geograpical coordinate of Karnataka are {}, {}.'.format(lat, lon))

except GeocoderTimedOut as e:
    print("Error: geocode failed on input %s with message %s"%(address, e.message))
```

# Getting the co-ordinates of individual districts in Karnataka

```python
address = df['District']
for state in address:
    geoloc = Nominatim(user_agent="corona")
    try:
        loc = geoloc.geocode(state, timeout=10)
        latitude = loc.latitude
        #if state == df_coords['District']:
            #df_coords['Latitude'] = latitude
        longitude = loc.longitude
        #df_coords['Longitude'].append(longitude)
        print('The geograpical coordinate of' + ' ' + state + ' '+ 'are {}, {}.'.format(latitude, longitude))
        df_coords = df_coords.append({'District': state, 'Latitude': latitude, 'Longitude': longitude}, ignore_index=True)

    except GeocoderTimedOut as e:
        print("Error: geocode failed on input %s with message %s"%(state, e.message))
```

# Getting venues using foursquare

```python
radius = 500
LIMIT = 100

venues = []

for lat, lng, district, active in zip(df_final['Latitude'], df_final['Longitude'], df_final['District'], df_final['Total active cases']):
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        lng,
        radius, 
        LIMIT)
    
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    for venue in results:
        venues.append((
            district, 
            active,
            lat, 
            lng, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
```

# Adding venue list to a dataframe

```python
venues = pd.DataFrame(venues)

venues.columns = ['District', 'Total active cases', 'DistrictLatitude', 'DistrictLongitude', 'VenueName', 'VenueLatitude', 'VenueLongitude', 'VenueCategory']

print(venues.shape)
venues.head()
```

# One hot encoding venue data

```python
# one hot encoding
kar_onehot = pd.get_dummies(venues[['VenueCategory']], prefix="", prefix_sep="")

# add district and active cases column back to dataframe
kar_onehot['District'] = venues['District'] 
kar_onehot['Total active cases'] = venues['Total active cases'] 

# move district and active cases column to the first column
fixed_columns = list(kar_onehot.columns[-2:]) + list(kar_onehot.columns[:-2])
kar_onehot = kar_onehot[fixed_columns]

print(kar_onehot.shape)
kar_onehot
```

# Grouping one hot encoded data by district and total cases

```python
kar_grouped = kar_onehot.groupby(["District", "Total active cases"]).mean().reset_index()
kar_grouped
```
